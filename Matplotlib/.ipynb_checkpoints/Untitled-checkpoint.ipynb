```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot()
```

```python
x = [1,2,3,4]
y = [11,22,33,44]
plt.plot(x,y)
```

```python
# 1st method
fig = plt.figure()
ax = fig.add_subplot()
plt.show
```

```python
# 2nd method
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

```python
# 3rd method (recommended)
fig, ax = plt.subplots()
ax.plot(x, y); # add some data
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
fig, ax = plt.subplots()
ax.plot(x, x**2); # add some data
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
nut_butter_prices = {"Almond Butter": 10,
                    "Peanut Butter": 8,
                    "Cashew Butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Butter Prices", ylabel="Price ($)", xlabel="Butters")
```

```python
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
```
